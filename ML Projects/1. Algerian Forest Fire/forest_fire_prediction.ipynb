```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset.csv",header=1,)
df.head()
```

```python
df.info()
```

## Data cleaning

```python
# cheching null objects
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.loc[122]
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
# Remove the null values
df= df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[122]
```

```python
df.columns
```

```python
# Fix spaces in column name
df.columns= df.columns.str.strip()
df.columns
```

```python
# Changing the datatypes of the required columns
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# Changing other columns datatypes
## GENERIC METHOD ##

objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != "Classes":
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Saving the cleaned dataset
df.to_csv("Algerian_forest_fires_CLEANED.csv", index=False)
```

## Exploratory data analysis

```python
# Drop the unnecessary data
df_copy = df.drop(['day','month', 'year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# Data encoding
df_copy['Classes']= np.where(df_copy['Classes'].str.contains("not fire"),0,1)

# (df_copy['Classes']=="not fire",0,1) it means that where classes in not fire replace it with o else replace it with 1
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# Plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percent = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting pie chart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percent,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of classes')
plt.show()
```
